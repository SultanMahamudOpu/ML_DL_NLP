```python
# creating vectors
import numpy as np
```

```python
# two tiny 3D vectors
a=np.array([2,1,3])
b=np.array([1,3,3])
print(a)
print(b)
```

```python
# vector operations (addition and subtraction)

add_ab = a+b
sub_ab = a-b

print(add_ab)
print(sub_ab)
```

```python
# Dot product (similarity of direction)
# positive হলে একই দিকে, 0 এর কাছাকাছি হলে perpendicular, negative হলে বিপরীত দিকে
#manual dot product: sum of pairwise products

dot_manual=a[0]*b[0]+a[1]*b[1]+a[2]*b[2]

dot_np=np.dot(a,b) # বা a@b
print(dot_manual)
print(dot_np)
```

```python
# Norms (Length orr Magnitude of vector)
# L2 norm, the usual length
l2_a=np.linalg.norm(a)
print(l2_a)
l1_a=np.linalg.norm(a,ord=1)
print(l1_a)
```

```python
# Normalizing a vectorr (unit vector)

unit_a = a/np.linalg.norm(a)
len_unit_a=np.linalg.norm(unit_a)
print(unit_a)
print(len_unit_a)
```

# stable or Outlier না থাকলে খুব ভালো কাজ করে
# value কে 0-1 এর মধ্যে নিয়ে আসে


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    "h":[150,160,170,180,190],
    "w":[50,60,70,80,90]
})
df
```

```python
mn=df.min()
mx=df.max()
rg=mx-mn

#print করে দেখার জন্য
mn,mx,rg
```

```python
ss= df-mn
ss.head()
```

```python
mm=ss/rg
mm.round(2)
```

```python
mm.plot(kind="bar")
plt.title("Min-Max Scaling [0-1]")
plt.xlabel("Indx")
plt.ylabel("Min-max value")
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({
    "id":[1,2,3,4],
    "color":["red","blue","green","red"],
    "size":["small","medium","large","medium"],
    "price":[10,12,15,11]
})
df
```

```python
d_color=pd.get_dummies(df["color"],prefix="colum_name",dtype=int)
```

```python
df_encoded=pd.concat([df,d_color],axis=1)
```

```python
df_encoded=df_encoded.drop("color",axis=1)
print(df_encoded)
```

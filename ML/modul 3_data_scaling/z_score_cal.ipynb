# symmetric or bell curve তখন z score

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({
    'h':[150,160,170,180,190],
    'w':[50,60,70,80,90]
})
df
```

```python
m=df.mean()
print('Mean:\n',m)
s=df.std()
print('Standart divation:\n',s)
```

```python
# z-score
z=(df-m)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title('Z scaling features')
plt.xlabel("Index")
plt.ylabel("Z value")
plt.show()
```

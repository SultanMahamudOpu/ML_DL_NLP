```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_squared_error, r2_score
```

```python
hours= np.array([2, 4, 6, 8, 10]).reshape(-1,1)
marks= np.array([55, 60, 70, 80, 88])

hours, marks
```

```python
model= LinearRegression()
model.fit(hours, marks)
pred= model.predict(hours)

print("m: ",model.coef_[0])
print("c: ",model.intercept_)
print("accuracy: ", pred)
```

```python
# compute Metrics
mae= mean_squared_error(marks, pred)
mse= mean_squared_error(marks, pred)
rmse= np.sqrt(mse)
r2= r2_score(marks, pred)
```

```python
print("MAE: ", mae)
print("MSE: ", mse)
print("RMSE: ", rmse)
print("R^2: ", r2)
```

```python
plt.scatter(hours, marks, label="Actual Data")

plt.plot(hours, pred, color="red", label= "Predicted Line")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Studey Hours vs Marks -Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

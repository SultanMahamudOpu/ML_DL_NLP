```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
```

```python
hours= np.array([2, 4, 6, 8, 10]).reshape(-1,1)
marks= np.array([55, 60, 70, 80, 88])

hours, marks
```

```python
lin_reg= LinearRegression()
lin_reg.fit(hours, marks)

print("m: ",lin_reg.coef_[0])
print("c: ",lin_reg.intercept_)
```

```python
# compute the cost (Mean squared error) MSE

pred_marks= lin_reg.predict(hours)
mse_lin= mean_squared_error(marks, pred_marks)

print("MSE for LinearRegression: ",mse_lin)
```

```python
plt.scatter(hours, marks, label="Actual Data")

plt.plot(hours, pred_marks, label="Linear Regression")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Studey Hours vs Marks -Linear Regression")
plt.legend()
plt.grid(True)
plt.show()
```

# Gradient Descent based model: SGDRegressor

```python
sgd_reg= SGDRegressor(
    max_iter= 1000,
    learning_rate= "invscaling",
    eta0= 0.01,
    random_state= 42
)

sgd_reg.fit(hours, marks)
print("m: ", sgd_reg.coef_[0])
print("c: ",sgd_reg.intercept_[0])


pred_sgd=sgd_reg.predict(hours)
mse_sgd= mean_squared_error(marks, pred_sgd)

print("MSE of SGDRegressor: ", mse_sgd)
```

```python
from matplotlib.lines import lineStyles
plt.scatter(hours, marks, label="Actual Data")

# plt.plot(hours, pred_marks, label="Linear Regression", linestyle="--")
# plt.plot(hours, pred_sgd, label="SGD Regression", linestyle=".-")
plt.plot(hours, pred_marks, linestyle="--", color="red", label="Linear Regression")
plt.plot(hours, pred_sgd, linestyle="-.", color="green", label="SGD Regression")

plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Linear Regression vs SGDRegressor")
plt.legend()
plt.grid(True)
plt.show()
```

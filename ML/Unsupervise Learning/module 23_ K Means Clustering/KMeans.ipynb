```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv")
df.head()
```

```python
df.info()
```

```python
X=df[['sepal_length', 'sepal_width']]
X.head()
```

```python
scaler= StandardScaler()
X_scaled= scaler.fit_transform(X)
```

```python
# selectiong value of k
k= 3
model= KMeans(n_clusters=k, init='k-means++', max_iter=300, random_state=42)
model.fit(X_scaled)
```

```python
# কি laels দিছে সেটা দেখতে চাইলে
labels= model.labels_
labels[:75]  # প্রথম 75 টা
```

```python
df['cluster']=labels # dataframe এ label গুলা একটা নতুন column হিসাবে add করে  আসল class এর সাথে তুলনা করে দেখি কেমন দিছে
df.head(150)
```

```python
plt.Figure(figsize=(10, 6))
sns.scatterplot(
    x=df['sepal_length'],
    y=df['sepal_width'],
    hue=df['cluster'],
    palette='viridis'
)
plt.title('K-Means clustering result')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()
```

```python
# finding optimal k (Elbow Method)
inertia_values= []
k_range= range(1, 11)

for k in k_range:
    km_model= KMeans(n_clusters=k, init='k-means++', random_state=42)
    km_model.fit(X_scaled)
    inertia_values.append(km_model.inertia_)
```

```python
plt.Figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python
# Evaluation with Silhouette Score
score= silhouette_score(X_scaled, labels)
score
```

Interpretation guide: 
- Above 0.5 → strong clustering
- Around 0.2–0.4 → overlapping clusters
- Near 0 → useless clustering

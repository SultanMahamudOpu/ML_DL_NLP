# Question 01: [ Marks 10 ]

What is a perceptron?
Explain its three main components and their roles.

**Write** Answer 01:
Perceptron is the simplest types of artificial neural networks that takes multiple input values and produces a single output.
A perceptron has three main components:


1.   Input Features<br>
    These are the input variables or features of the data.
2.   Weights<br>
    Each input is associated with a weight that represents its importance in the decision-making process.
3. Bias<br>
  The bias helps us to shift the decision boundary







# Question 02: [ Marks 10 ]

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.

## Write Answer 02:
A decision boundary is a line, plan, or hyperplane that separates different classes.<br>
The equation:
w₁x₁ + w₂x₂ + b = 0<br>
This equation represents a straight line in a 2D space. This line divides the space into two regions:



1.   One side where the output is positive (class 1)
2.   Another side where the output is negative (class 0)

this equation defines the boundary where the perceptron is uncertain. Points lying exactly on the line satisfy the equation, while points on different sides belong to different classes.

# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:
A perceptron is called a linear classifier because it can only separate data using a linear decision boundary.

This is related to the concept of linear separability, where data points can be separated using a straight line.

Examples:


1.   AND Gate: Linearly separable
2.   OR Gate: Linearly separable
3.   XOR Gate: Not linearly separable


For AND and OR, we can draw a straight line that separates output classes. But for XOR, no single straight line can separate the classes correctly.

Where:

n = learning rate

y = actual output

ynot = predicted output

W = Weights

B =Bais

**Intuition:**


1.   If the prediction is correct no change is needed
2.   If the prediction is wrong weights are adjusted to reduce error

Weights are updated only when the prediction is incorrect because:


1.   Correct predictions mean the model is already performing well for that data point
2.   Updating weights unnecessarily may disturb already correct classifications


# Question 05: [ Marks 10 ]

Why does a single-layer perceptron fail to solve the XOR problem?
What does this limitation tell us about the need for multilayer neural networks?

## Write Answer 05:

A single layer perceptron fails to solve the XOR problem because XOR is not linearly separable.

In XOR:


1.   Output is 1 when inputs are different
2.   Output is 0 when inputs are same

These points cannot be separated using a single straight line.


This limitation shows that:


1.   Single-layer perceptrons are too simple
2.   We need multi-layer neural networks to handle complex patterns




# Question 06: [ Marks 20 ]

What is meant by the “perceptron Learning Rule” or weight adjustment method?
Why can adjusting one weight affect the classification of other points?

## Write Answer 06:

The perceptron learning rule is a method for updating weights based on prediction error.

This means:


1.   Increase weight if prediction is too low
2.   Decrease weight if prediction is too high


Why changing one weight affects others:

All inputs contribute to the final output through the weighted sum. So:


1.   Changing one weight changes the decision boundary
2.   This may shift classification of other data points

What are the main limitations of a single-layer perceptron?
How does the idea of using more than one layer (MLP) help overcome these limitations at a high level accoriding to you?



## Write Answer 07:

Some limitations of a single-layer perceptron:


1.   Can only solve linearly separable problems
2.   Cannot solve XOR-type problems
3.   Limited representation power
4. Uses a simple step activation function


To overcome these limitations, we use Multi-Layer Perceptron (MLP):


1.   MLP introduces hidden layers
2.   Applies non-linear activation functions
3. Can learn complex patterns and relationships

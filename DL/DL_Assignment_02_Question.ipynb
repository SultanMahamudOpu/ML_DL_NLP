# Question 01: [ Marks 10 ]
Explain gradient descent in detail. What is the main idea behind it? How does it help in minimizing the loss function? Explain the role of learning rate and describe what happens if the learning rate is too high or too low.

# **Write** Answer 01:

Gradient Descent is an optimization algorithm used to minimize the loss function in machine learning models. The main idea is to iteratively adjust the weights in the direction that reduces the error.

It works by calculating the gradient (slope) of the loss function with respect to the parameters and then updating the parameters in the opposite direction of the gradient. This is because the gradient points toward the steepest increase, so moving in the opposite direction decreases the loss.

The update rule is:

$$
\theta = \theta - \alpha \cdot \nabla J(\theta)
$$

where:
- $\theta$ = parameters  
- $\alpha$ = learning rate  
- $\nabla J(\theta)$ = gradient of loss  

### Role of Learning Rate:
- If learning rate is too high → model may overshoot and fail to converge  
- If learning rate is too low → training becomes very slow and may get stuck  


# Question 02: [ Marks 10 ]

What is forward propagation in an artificial neural network? Explain step by step how data moves from the input layer to the output layer.

## Write Answer 02:

Forward propagation is the process of passing input data through the neural network to generate an output.

### Steps:
1. Input features are fed into the input layer  
2. Each input is multiplied by weights and added with bias  
3. The result is passed through an activation function  
4. This process continues through hidden layers  
5. Finally, the output layer produces the prediction  

### Mathematical Representation:

$$
Z = WX + b
$$

$$
A = f(Z)
$$

This process continues layer by layer until the final output is obtained.

# Question 03: [ Marks 15 ]
Why do we need activation functions in neural networks? What would happen if we only used linear transformations without any activation function?

## Write Answer 03:

## Need for Activation Functions

Activation functions are essential in neural networks because they introduce non-linearity into the model.

Without activation functions:
- The network would only perform linear transformations  
- No matter how many layers are added, the output would still be a linear combination of the input  
- The model would behave like a simple linear regression  

Mathematically, multiple linear layers collapse into one:

$$
Y = WX + b
$$

So, the network cannot learn complex patterns such as images, speech, or non-linear relationships.

# Question 04: [ Marks 15 ]

Explain the sigmoid activation function. Where is it used? What are its advantages and disadvantages? What problem does it cause during training?

## Write Answer 04:

The sigmoid activation function is a mathematical function that maps input values into a range between 0 and 1.

It is defined as:

$$
\sigma(x) = \frac{1}{1 + e^{-x}}
$$

### Where is it used
- Mainly used in the output layer for binary classification problems
- Outputs can be interpreted as probabilities (between 0 and 1)

### Advantages:
- Smooth and differentiable  
- Output range (0,1) makes it suitable for probability interpretation  
- Easy to understand and implement  

### Disadvantages:
- Suffers from vanishing gradient problem
- Output is not zero-centered  
- Slow convergence during training  

### Problem during training:
For very large positive or negative input values, the gradient becomes very small (close to zero).  
As a result, weight updates become extremely slow, which is known as the vanishing gradient problem.

# Question 05: [ Marks 10 ]
Explain the softmax activation function in simple words. Why is it used in multi-class classification problems? How is it different from sigmoid?

## Write Answer 05:

The Softmax activation function converts a set of values into probabilities.

It takes multiple input values and transforms them into outputs between 0 and 1, where the sum of all outputs equals 1.

The formula is:

$$
\text{Softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}
$$

### Why is it used in multi-class classification?
- In multi-class problems, we need to assign probabilities to multiple classes  
- Softmax ensures that:
  - Each output represents the probability of a class  
  - All probabilities add up to 1  
- This makes it easy to choose the class with the highest probability as the final prediction  

### Difference between Softmax and Sigmoid:
- **Sigmoid:**
  - Used for binary classification  
  - Outputs a single probability

- **Softmax:**
  - Used for multi-class classification  
  - Outputs multiple probabilities that depend on each other  
  - Ensures total probability = 1  


# Question 06: [ Marks 20 ]

Explain the structure of an Artificial Neural Network in detail. Describe the role of the input layer, hidden layers, and output layer. How do weights, biases, and activation functions work together to transform input data into predictions? Illustrate your explanation with a simple example.

## Write Answer 06:

An Artificial Neural Network (ANN) is a computational model inspired by the human brain. It consists of multiple layers that work together to transform input data into meaningful predictions.

### 1. Input Layer
- The input layer receives the raw data (features)  
- Each neuron represents one feature of the dataset  
- It does not perform any computation, only passes data to the next layer  

### 2. Hidden Layers
- Hidden layers perform the main computations  
- Each neuron takes input, multiplies it by weights, adds bias, and applies an activation function  
- These layers help the network learn complex patterns and relationships in the data  

### 3. Output Layer
- The output layer produces the final prediction  
- The type of activation function depends on the task:
  - Sigmoid → binary classification  
  - Softmax → multi-class classification   


### Mathematical Representation

At each neuron:

$$
Z = WX + b
$$

$$
A = f(Z)
$$

Where:
- $W$ = weights  
- $X$ = input  
- $b$ = bias  
- $f$ = activation function  
- $A$ = output  


### How Components Work Together

- **Weights ($W$):** Control the importance of each input feature  
- **Bias ($b$):** Helps shift the output, allowing better fitting  
- **Activation Function ($f$):** Introduces non-linearity so the network can learn complex patterns  

The process:
1. Input data enters the network  
2. Each layer computes $Z = WX + b$  
3. Activation function is applied → $A = f(Z)$  
4. Output is passed to the next layer  
5. Final layer produces prediction  


### Simple Example

Suppose we want to predict whether a student will pass or fail.

- **Input Layer:**
  - Study hours  
  - Attendance  

- **Hidden Layer:**
  - Learns relationship between study habits and performance  

- **Output Layer:**
  - Predicts: Pass (1) or Fail (0)  

Example computation:

$$
Z = (w_1 \cdot \text{Study Hours}) + (w_2 \cdot \text{Attendance}) + b
$$

$$
A = \sigma(Z)
$$

If $A > 0.5$ → Pass  
If $A \leq 0.5$ → Fail

```python
import pandas as pd
import numpy as np
import torch
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df= pd.read_csv('diabetes.csv')
df
```

```python
print(df.isnull().sum())
```

```python
X = df.drop('Outcome', axis=1).values
y = df['Outcome'].values
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_np = np.array(X_scaled)
y_np = np.array(y)

X_tensor = torch.tensor(X_np, dtype=torch.float32)
y_tensor = torch.tensor(y_np, dtype=torch.float32).view(-1, 1)

X_train, X_test, y_train, y_test = train_test_split(
    X_tensor, y_tensor, test_size=0.2, random_state=42
)

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Design a neural network using nn.Module.

## Write Answer 02:


```python
import torch
import torch.nn as nn
```

```python
class DiabetesModel(nn.Module):
    def __init__(self):
        super(DiabetesModel, self).__init__()
        self.fc1 = nn.Linear(8, 16)

        self.relu = nn.ReLU()

        self.fc2 = nn.Linear(16, 1)

        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        x = self.sigmoid(x)
        return x

model = DiabetesModel()

print(model)

total_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
print("Total Trainable Parameters:", total_params)
```

## 1. Input Layer

The dataset has **8 input features**, so the input layer contains:

$$
text{Input neurons} = 8
$$

**Example features:**
- Pregnancies  
- Glucose  
- BloodPressure

- etc.


## 2. Hidden Layer

I used **16 hidden neurons**.

$$
text{Hidden neurons} = 16
$$

**Justification:**
- More neurons allow the model to learn complex patterns  
- 16 provides a good balance between learning capacity and avoiding overfitting  
- Suitable for small to medium datasets  

**Network structure:**
$$
8 \rightarrow 16 \rightarrow 1
$$


## 3. Output Layer

The output layer has:

$$
text{Output neurons} = 1
$$

**Reason:**
- This is a binary classification problem  
- Output classes:
  - \(0\) = No Diabetes  
  - \(1\) = Diabetes  


## 4. Choice of Activation Function

### Hidden Layer: ReLU

$$
text{ReLU}(x) = \max(0, x)
$$

**Why ReLU?**
- Computationally efficient  
- Helps reduce vanishing gradient problem  
- Faster training  


### Output Layer: Sigmoid

$$
sigma(x) = \frac{1}{1 + e^{-x}}
$$

**Why Sigmoid?**
- Converts output into probability (0 to 1)  
- Suitable for binary classification  


## 5. Trainable Parameters Calculation

### Layer 1 (Input → Hidden)

$$
(8 \times 16) + 16 = 144
$$


### Layer 2 (Hidden → Output)

$$
(16 \times 1) + 1 = 17
$$


## Total Trainable Parameters

$$
144 + 17 = 161
$$

# Question 04: [ Marks 15 ]

## Implement a full training loop:

Forward pass

Loss computation

Backward pass

Parameter update

Gradient reset

### Requirements:

Train for at least 100 epochs.

Print loss every 10 epochs.

Store training loss history(You can pick your own Data Structure).

Explain clearly what happens in each step of the pipeline.

## Write Answer 04:

```python
epochs = 100
loss_history = []

for epoch in range(epochs):
    outputs = model(X_train)

    loss = criterion(outputs, y_train)

    optimizer.zero_grad()
    loss.backward()

    optimizer.step()

    loss_history.append(loss.item())

    if (epoch+1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")
```


Forward pass → prediction.

Loss computation → compare prediction with actual.

Backward pass → calculate gradients.

Optimizer step → update weights.

Zero gradients → clear old gradients.

```python
with torch.no_grad():
    y_pred = model(X_test)
    predicted = (y_pred > 0.5).float()

    accuracy = (predicted == y_test).float().mean()
    print("Test Accuracy:", accuracy.item()*100)
```

The model is evaluated using accuracy, as it is a classification problem.

The predictions are converted into binary values using a threshold of 0.5.



If training accuracy is much higher than testing accuracy, the model is overfitting.

If both are low, the model is underfitting.

## Write Answer 07:

1. Why reset gradients every epoch?

PyTorch accumulates gradients by default. Resetting prevents adding old gradients.

2. What happens if learning rate is too high?

Loss may oscillate or diverge.

3. What happens if learning rate is too small?

Training becomes very slow and may get stuck.

4. Why define layers inside __init__() not in forward()?

Layers need to be initialized once and their parameters tracked by PyTorch.

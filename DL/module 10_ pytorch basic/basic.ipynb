````python
m2= m2.T 
m2
````

````python
torch.matmul(m1,m2)
````

# Inplace operations
```
 অন্য variable এর মধ্যে রাখলে জায়গার নষ্ট হবে, যখন বড় architer নিয়ে কাজ করব তখন full code crash করবে। তাই  জায়গা বাচানোর জন্য ওটার মধ্যে update করে রেখে দিব, এটাই inplace o perations
```

````python
a= torch.tensor([1, 2, 3])
a
````

````python
b= a+ 5
b
````

````python
m= torch.rand(2, 3)
n= torch.rand(2, 3)
````

````python
m
````

````python
n
````

````python
# in place operation

m.add_(n) # m টা এখানে update হয়ে গেছে, আলাদা variable লাগবে না
````

````python
torch.relu_(m) # সব কিছুই inplace version আছে, নিচে underscore মানেই inplace
````

````python
n.relu_() # এভাবেও করা যায়
````

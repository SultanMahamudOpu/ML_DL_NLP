# Tensors Shape

````python
x= torch.tensor([[1,2,3], [4,5,6]])
x
````

````python
x.shape
````

````python
torch.empty_like(x) # x এর shape এর মত, empty tensor বানবে
````

````python
torch.ones_like(x) # x এর shape এর মত, one নিয়ে tensor বানবে
````

````python
torch.rand_like(x, dtype=torch.float32) # x এর shape এর মত, random (float বলে দিছি, অন্য কিছুও বলা যাইত) নিয়ে tensor বানবে
````

# Data types

````python
x.dtype
````

````python
torch.tensor([1.0, 2.3, 3.0, 4.5], dtype=torch.int32) # data টা float কিন্তু আমরা চাচ্ছি int এ
````

````python
torch.tensor([1, 2, 4, 5], dtype=torch.float32) # data টা int কিন্তু আমরা চাচ্ছি float এ
````

````python
print('প্রথমে এটা int ছিল: ',x.dtype)
x=x.to(torch.float32)
print(x)
print('convert করে এখন এটাকে flot করা হল: ',x.dtype)
````
